```python
from transformers import AutoTokenizer
checkpoint = 'bert-base-cased'
tokenizer = AutoTokenizer.from_pretrained(checkpoint)
print(tokenizer)
```

```python
result = tokenizer("Hello Learner")
print(result)
```

```python
tokens = tokenizer.tokenize("Hello Learner")
print("tokens : ",tokens)
```

```python
full_ids = tokenizer.encode("Hello Learner")
print("full_ids : ",full_ids)
print("decoded tokens :",tokenizer.convert_ids_to_tokens(full_ids))
```

```python
vocab = tokenizer.get_vocab()
print(vocab)
```

```python
print(f"Vocabulary Size :{len(vocab)}")
print(f"Hello : {vocab['hello']}")
print(f"Learner : {vocab['Lea']}")
print(f"Learner : {vocab['##rner']}")
```
